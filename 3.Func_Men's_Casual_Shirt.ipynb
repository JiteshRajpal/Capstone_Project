```python
def create_mrp_range_column(input_df):
    """
    Creates a new column 'mrp_range' by binning the 'mrp' column into specified ranges.

    Args:
    - input_df (pd.DataFrame): The input DataFrame to be processed.

    Returns:
    - pd.DataFrame: The processed DataFrame with the new 'mrp_range' column.
    """
    # Create bin edges and labels
    bins = [0, 1000, 2000, 3000, 4000]  # Extra bin edge for values > 3000
    labels = ['Low Price Range', 'Mid Price Range', 'High Price Range', 'Premium Range']

    # Apply binning to the 'mrp' column and create the new column 'mrp_range'
    input_df['mrp_range'] = pd.cut(input_df['mrp'], bins=bins, labels=labels)

    return input_df
```

```python
def remove_na_values(input_df):
    while input_df.isna().any().any():
        # Find the column with the highest number of NaN values
        col_with_max_na = input_df.isna().sum().idxmax()

        # Remove all rows containing NaN values in the entire DataFrame
        input_df.dropna(subset=[col_with_max_na], inplace=True)
```

```python
from kmodes.kmodes import KModes
```

```python
def elbow_curve(input_df, num_clusters_range, init_method="Huang", n_init=5, verbose=1):
    cost = []
    
    for num_clusters in num_clusters_range:
        kmode = KModes(n_clusters=num_clusters, init=init_method, n_init=n_init, verbose=verbose)
        kmode.fit_predict(input_df)
        cost.append(kmode.cost_)

    # Plotting the elbow method
    plt.plot(num_clusters_range, cost, 'bx-')
    plt.xlabel('No. of clusters')
    plt.ylabel('Cost')
    plt.title('Elbow Method For Optimal K')
    plt.show()
```

```python
elbow_curve(train_set, num_clusters_range=range(1, 50))
```

```python
def perform_kmodes_clustering(input_df, n_clusters, init_method="random", n_init=5, verbose=1):
    kmode = KModes(n_clusters=n_clusters, init=init_method, n_init=n_init, verbose=verbose)
    clusters = kmode.fit_predict(input_df)

    # Check if 'Clusters' column exists before dropping
    if 'Clusters' in input_df.columns:
        input_df.drop('Clusters', axis=1, inplace=True)

    # Insert 'Clusters' column with cluster labels at the beginning
    input_df.insert(0, "Clusters", clusters, True)

    return input_df, kmode  # Returning both the DataFrame and the trained KModes model
```

```python
train_set, kmode_model = perform_kmodes_clustering(train_set, n_clusters=15)
```

```python
def perform_kmodes_predictions(input_df, kmode_model):
    predictions = kmode_model.predict(input_df)

    # Check if 'Prediction' column exists before dropping
    if 'Prediction' in input_df.columns:
        input_df.drop('Prediction', axis=1, inplace=True)

    # Insert 'Prediction' column with prediction labels at the beginning
    input_df.insert(0, "Prediction", predictions, True)
```

```python
from sklearn.metrics import jaccard_score
```

```python
def find_best_matches(test_df, result_df):
    # Initialize an empty list to store the best match for each row
    best_matches = []

    # Iterate over each row in the test_df
    for _, row in test_df.iterrows():
        # Calculate Jaccard similarities for each row in the result_df
        similarities = result_df.apply(lambda x: jaccard_score(row[1:], x[1:], average='weighted'), axis=1)

        # Get the index and similarity score of the best match
        best_match_index = similarities.idxmax()
        best_match_similarity = similarities.max()

        # If the maximum similarity is greater than 0 (meaning a match exists)
        if best_match_similarity > 0:
            # Get the values for the best match and append to the result list
            best_matches.append({
                'Prediction': row['Prediction'],
                'Clusters': result_df.loc[best_match_index, 'Clusters'],
                'Similarity': best_match_similarity,
                'item_pattern': result_df.loc[best_match_index, 'item_pattern'],
                'item_primary_color': result_df.loc[best_match_index, 'item_primary_color'],
                'item_primary_material': result_df.loc[best_match_index, 'item_primary_material'],
                'item_product_fit': result_df.loc[best_match_index, 'item_product_fit'],
                'item_neck_type': result_df.loc[best_match_index, 'item_neck_type'],
                'item_sleeve_length': result_df.loc[best_match_index, 'item_sleeve_length'],
                'item_topwear_length': result_df.loc[best_match_index, 'item_topwear_length'],
                'mrp_range': result_df.loc[best_match_index, 'mrp_range']
            })
        else:
            # If no match is found, append a placeholder row with a similarity of 0
            best_matches.append({
                'Prediction': row['Prediction'],
                'Clusters': None,  # Or some other placeholder value
                'Similarity': 0,
                'item_pattern': None,
                'item_primary_color': None,
                'item_primary_material': None,
                'item_product_fit': None,
                'item_neck_type': None,
                'item_sleeve_length': None,
                'item_topwear_length': None,
                'mrp_range': None
            })

    # Convert the list of best matches to a DataFrame
    best_matches_df = pd.DataFrame(best_matches)

    return best_matches_df
```

```python
def merge_dataframes_on_attributes(CS_Men, best_matches_df):
    # Define the common attributes for merging
    common_attributes = ['item_pattern', 'item_primary_color', 'item_primary_material', 'item_product_fit',
                         'item_neck_type', 'item_sleeve_length', 'item_topwear_length', 'mrp_range']

    # Merge the DataFrames on the specified attributes, keeping only rows from CS_Men
    common_rows = CS_Men.merge(best_matches_df, on=common_attributes, how='inner')

    return common_rows
```

```python
common_rows=merge_dataframes_on_attributes(CS_Men,best_matches_df)
```

```python
def filter_rows_by_days(df, days_threshold):
    # Filter rows based on the condition: days_since_introduction < days_threshold
    filtered_df = df[df['days_since_introduction'] < days_threshold]

    return filtered_df
```

```python
less_than_60 = filter_rows_by_days(common_rows,60)
```

```python
def add_unique_item_count_column(input_df):
    # Add a new column with the unique item count per day since introduction
    input_df['unique_item_count_per_day_since_introduction'] = input_df.groupby('days_since_introduction')['item_id'].transform('nunique')

    return input_df
```

```python
def add_days_bucket_column_inplace(input_df):
    # Create bins for days_since_introduction
    bins = [0, 7, 14, 21, 28, 35, 42, 49, 56, float('inf')]  # Adjust the bin edges as needed
    labels = ['1-7', '8-14', '15-21', '22-28', '29-35', '36-42', '43-49', '50-56', '57+']

    # Add a new column with the bucket labels to the original DataFrame
    input_df['days_bucket'] = pd.cut(input_df['days_since_introduction'], bins=bins, labels=labels, right=False)
```

```python
def create_pivot_table_1(input_df):
    pivot_table = pd.pivot_table(
        input_df,
        values='qty',
        index=['item_id', 'l1_category', 'l0_category', 'l2_category', 'item_colour', 'item_pattern', 'item_primary_color', 'item_primary_material', 'item_product_fit', 'item_occasion', 'item_gender', 'item_neck_type', 'item_sleeve_length', 'item_topwear_length', 'item_brand_name'],
        columns='days_bucket',
        aggfunc='sum',
        fill_value=0
    ).reset_index()

    return pivot_table
```
